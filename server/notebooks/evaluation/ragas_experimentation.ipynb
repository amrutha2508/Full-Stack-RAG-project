```python
# answer relevancy 
# Does the answer generated actually answer what the user asked

from ragas.embeddings.base import embedding_factory
from ragas.metrics.collections import AnswerRelevancy

# Setup LLM and embeddings

embeddings = embedding_factory("openai", model="text-embedding-3-small", client=client)

# Create metric
scorer = AnswerRelevancy(llm=llm, embeddings=embeddings)

# Evaluate
result3 = await scorer.ascore(
    user_input = "How do I determine the correct size for the CM-7M?",
    response="""
    To determine the correct size for the CM-7M® gas mask, you need to measure your facial dimensions, specifically the facial height (from the nose bridge to the chin) and the width of the face (bizygomatic width). The CM-7M® comes in three sizes, and the size chart is as follows:

        - **Size 1 (Small):** Facial height of 108 – 120 mm and width of 125 – 140 mm.
        - **Size 2 (Medium):** Facial height of 120 – 128 mm and width of 140 – 148 mm. This size fits 95% of the population and is recommended for average-sized individuals.
        - **Size 3 (Large):** Facial height of 128 – 138 mm and width of 148 – 156 mm. This size is suggested for users over 250 lbs with wide faces.

        To ensure the mask fits correctly, perform a negative pressure test by blocking the inhalation valve and taking a deep breath. The mask should seal tightly to your face without air leakage.
    """
)
print(f"Answer Relevancy Score: {result3.value}")
```
